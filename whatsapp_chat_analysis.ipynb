```python
import pandas as pd
import re
import numpy as np
```

```python
f = open('WhatsApp Chat with Mayank Dsmnru.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
User_message = r"(\d{2}/\d{2}/\d{4}),(\d{1,2}:\d{2}\s(?:am|pm))- ([^:]+): (.+)"
System_message = r"(\d{2}/\d{2}/\d{4}),(\d{1,2}:\d{2}\s(?:am|pm)) - (.+)"
```

```python
#lists to store sparse data
Dates , Times , Authors , messages = [] , [] , [] ,[]
for line in data.split('\n'):
    match = re.match(User_message,line)
    if match:
        Date , Time , Author , message = match.group()
        Dates.append(Date)
        Times.append(Time)
        Authors.append(Author)
        messages.append(messages)
    else:
        match = re.match(System_message,line)
    if match:
        Date , Time , Author , message = match.group()
        Dates.append(Date)
        Times.append(Time)
        Authors.append('System')
        messages.append(messages)
df = pd.DataFrame({
    'Date':Dates,
    'Time':Times,
    'Author':Authors,
    'Message':messages
})
df.to_csv("Chat_Data.csv" ,index=False)
print("chat data is save 'Chat_Data.csv'.")

```
